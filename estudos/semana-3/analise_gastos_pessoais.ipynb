## 8. Seleção inicial de colunas

Para uma primeira análise, foram selecionadas as colunas `data`, `categoria` e `valor`.

Essas colunas permitem entender:

- quando o gasto ocorreu;
- em qual categoria o gasto foi registrado;
- qual foi o valor do gasto.

```python
df["valor"].sum()
```

## 9. Total de gastos

Foi calculado o total geral dos gastos registrados no dataset.

```python
df["valor"].mean()
```

## 10. Média dos gastos

Foi calculada a média dos valores gastos, permitindo ter uma noção do valor médio por lançamento.

```python
df["valor"].max()
```

## 11. Maior gasto

Foi identificado o maior valor de gasto registrado no dataset.

```python
df["valor"].min()
```

## 12. Menor gasto

Foi identificado o menor valor de gasto registrado no dataset.

```python
df["categoria"].value_counts()
```

## 13. Frequência por categoria

Foi verificada a quantidade de lançamentos por categoria, permitindo observar quais categorias aparecem com mais frequência no dataset.

```python
df.groupby("categoria")["valor"].sum()
```

## 14. Total gasto por categoria

Foi realizado um agrupamento por categoria para identificar o total gasto em cada uma delas.

Essa análise ajuda a entender quais categorias concentraram os maiores gastos.
